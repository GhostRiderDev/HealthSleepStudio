```python
import pandas as pd
```

```python
data = pd.read_csv('data/health_Sleep.csv', index_col='Person ID')
data["Sleep Disorder"] = data["Sleep Disorder"].fillna("None")
data.head()
```

#### ¿Quien tiene la peor calidad de sueño las mujeres o los hombres?


```python
# Print the data type of each column
data.info()
```

```python
# Print the summary statistics of the data
data.describe()
```

#### Analisis de generos


```python
data["Gender"].value_counts()
```

```python
data.loc[data["Age"] <= 40, "Age_group"] = "ADULT"
data.loc[data["Age"] >= 41, "Age_group"] = "OLDER ADULT"
```

```python
data.head()
```

```python
data.tail()
```

```python
data["Age_group"].value_counts()
```

```python
data.groupby("Gender")["Age"].agg(["mean", "std"])
```

#### Analisis de las ocupaciones


```python
data.groupby("Occupation")["Gender"].value_counts()
```

```python
data.groupby("Gender")[["Sleep Duration", "Quality of Sleep"]].mean()
```

```python
occu_qua = data.groupby("Occupation")[["Sleep Duration", "Quality of Sleep"]].mean()
occu_qua.sort_values(by="Quality of Sleep", ascending=False)
```

```python
# indice de masa corporal por ocupaciones
occu_bmi = data.groupby("Occupation")["BMI Category"].value_counts()
occu_bmi.sort_values()
```

```python
occu_steps = data.groupby("Occupation")[["Daily Steps", "Physical Activity Level", "Sleep Duration", "Quality of Sleep"]].mean()
occu_steps.sort_values("Daily Steps").head()
```
